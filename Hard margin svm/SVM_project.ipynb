```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from Hard_margin_svm import HardMarginSVM
from sklearn.svm import LinearSVC
```

```python
data=pd.read_csv("data.csv")
data
```

```python
X=data.drop(["class"] , axis=1)
y=data["class"]
y=np.where(y==0 , -1 , y)
```

```python
scaler=StandardScaler()
X_std=scaler.fit_transform(X)
```

```python
model=HardMarginSVM()
model.fit(X_std , y)
model.w
```

```python
plt.scatter(X_std[: , 0] , X_std[: , 1] , c=y)
xx1 , xx2 =np.meshgrid(
    np.arange(-2 , 2, 0.01),
    np.arange(-2 , 2 , 0.01)
)
meshgrid_points=np.array([xx1.ravel() , xx2.ravel()]).T
z=model.predict(meshgrid_points)
z=z.reshape(xx1.shape)
plt.contourf(xx1, xx2 ,z , alpha=.2)
```

```python
model=LinearSVC()
model.fit(X_std , y)
```

```python
plt.scatter(X_std[: , 0] , X_std[: , 1] , c=y)
xx1 , xx2=np.meshgrid(
    np.arange(-2 , 2, 0.01 ),
    np.arange(-2 , 2, 0.01)
)
meshgrid_points=np.array([xx1.ravel(), xx2.ravel()]).T
z=model.predict(meshgrid_points)
z=z.reshape(xx1.shape)
plt.scatter(xx1 , xx2 ,z , alpha=0.02)







```
